```python
west_countries = [9288,9780,9969,10239,10020,10533,10590,10588,10343,10888,10986,11051,11537,11872,12386,12997,13653,13846,14353,14692,13597,14849,15443,16172,16048,15886,16545,17064,17798,18228,18057,18341,17816,18334,19465,20063,20533,21067,21753,22254,22356]
```

```python
ussr = [2834,2798,2928,3004,3098,3304,2557,3566,3768,3660,3935,4088,4130,3976,4430,4626,4796,4955,5194,5218,5569,5663,5640,6058,6175,6136,6366,6459,6565,6480,6437,6442,6544,6692,6715,6715,6924,6943,7032,7078,6871]
```

```python
years = range(1950,1991)
```

```python
years
```

```python
plt.plot(years, west_countries, color = 'blue')
plt.ylabel('PIB per capita médio de Austrália, Canadá,\n Nova Zelândia e EUA em dólares')
plt.savefig('eua.png')
```

```python
plt.plot(years,ussr, color = 'yellow')
plt.ylabel('PIB per capita da URSS em dólares')
plt.savefig('ussr.png')
```

```python
pib_comparado = [x/y for x, y in zip(ussr, west_countries)]
pib_comparado = [element * 100 for element in pib_comparado]
```

```python
fig,ax = plt.subplots()
ax.yaxis.set_major_formatter(PercentFormatter())
plt.ylabel('% do PIB per Capita médio de\nAustrália, Canadá, Nova Zelândia e EUA')
plt.plot(years,pib_comparado, color = 'red')
plt.savefig('ussr-comparado.png')
```
